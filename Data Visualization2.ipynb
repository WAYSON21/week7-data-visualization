```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 df = pd.read_csv('iris.data')
```

```python
df
```

```python
print("\n First 5 rows of the dataset:")
print(df.head())
```

```python
print("\n Data types and missing values:")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
df_clean = df.dropna()
print(f"\n Cleaned dataset shape: {df_clean.shape}")
```

```python
print("\n Descriptive statistics:")
print(df_clean.describe())
```

```python
print("\n Mean values grouped by species:")
grouped_means = df_clean.groupby("species").mean()
print(grouped_means)
```

```python
print("\n Observations:")
print("- Virginica flowers tend to have the longest petals and sepals.")
print("- Setosa has shorter petal lengths and widths compared to others.")
print("- Versicolor is intermediate in most measurements.")
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
grouped_means["petal_length"].plot(marker='o')
plt.title("Average Petal Length by Species")
plt.ylabel("Petal Length (cm)")
plt.xlabel("Species")
plt.grid(True)
plt.tight_layout()
plt.savefig("line_chart_petal_length.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_clean, x="species", y="sepal_width", estimator="mean", palette="Set2")
plt.title("Average Sepal Width per Species")
plt.ylabel("Sepal Width (cm)")
plt.xlabel("Species")
plt.tight_layout()
plt.savefig("bar_chart_sepal_width.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df_clean, x="petal_length", kde=True, bins=20)
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("histogram_petal_length.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_clean, x="sepal_length", y="petal_length", hue="species", style="species")
plt.title("Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.tight_layout()
plt.savefig("scatter_sepal_vs_petal.png")
plt.show()
```
